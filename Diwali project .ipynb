```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Diwali Sales Data.csv',encoding= 'unicode_escape')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace= True)
pd.isnull(df).sum()
```

```python
df.dropna(inplace = True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns= {'Marital_Status': 'Shaadi'})
```

```python
df.describe()
```

```python
# for specific columns 
df[['Orders','Amount']].describe()
```

```python
# plotting a bar chart for Gender and it's count
a= sns.countplot(x='Gender', data=df)
for bars in a.containers :
    a.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount
sales_gen =df.groupby(['Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending= False)
```

```python
# plotting a bar chart for gender vs total amount
sns.barplot(x= 'Gender', y= 'Amount',data=sales_gen)
```

```python
k= sns.countplot(x='Age Group', hue = 'Gender', data= df)
for bars in k.containers :
    k.bar_label(bars)
    
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index= False)['Amount'].sum().sort_values('Amount', ascending= False)
sns.barplot(x='Age Group', y= 'Amount', data= sales_age)
```

```python
# total number of orders from top 10 states
sales_state= df.groupby('State',as_index= False)['Orders'].sum().sort_values('Orders', ascending= False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x= 'State', y='Orders', data= sales_state)
```

```python
# total amount/sales from top 10 states
sales_state= df.groupby('State', as_index= False)['Amount'].sum().sort_values('Amount', ascending= False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x= 'State', y= 'Amount', data= sales_state)
```

```python
# Marital Status
m= sns.countplot(x= 'Marital_Status', data= df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in m.containers :
        m.bar_label(bars)
 
```

```python
sales_state_1= df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values('Amount')
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status', y='Amount', data=sales_state_1, hue='Gender')
```

```python
#Occupation 
O=sns.countplot(x = 'Occupation', data=df)
sns.set(rc={'figure.figsize':(17,5)})
for bars in O.containers:
    O.bar_label(bars)
```

```python
sales_state= df.groupby('Occupation',as_index= False)['Amount'].sum().sort_values('Amount', ascending= False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'Occupation', y= 'Amount',data= sales_state)
```

```python
# Product Category
ax= sns.countplot(x='Product_Category', data= df)
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers :
    ax.bar_label(bars)
```

```python
sales_state= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values('Amount', ascending= False).head(10)
sns.barplot(x= 'Product_Category', y= 'Amount', data= sales_state)
sns.set(rc= {'figure.figsize':(30,5)})
```

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values('Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
#top 10 most sold products 
fig1 , ax1 = plt.subplots(figsize=(12,7))  
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
